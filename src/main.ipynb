```python
import pandas as pd
import requests

# Read in your symbols
sp_500 = ['AAPL']

# Endpoint
url = "https://query2.finance.yahoo.com/v1/finance/esgChart"

# List of dataframes
dataframes = []

for symbol in sp_500:
    response = requests.get(url, params={"symbol": symbol})
    if response.ok:
        df = pd.DataFrame(response.json()["esgChart"]["result"][0]["symbolSeries"])
        df["symbol"] = symbol
        dataframes.append(df)

df = pd.concat(dataframes)
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
```

```python
from yahooquery import Ticker

def get_esg_score(ticker_symbol):
    ticker = Ticker(ticker_symbol)
    esg_data = ticker.esg_scores

    if isinstance(esg_data, dict) and esg_data.get('esgPerformance', None) is not None:
        esg_score = esg_data['esgPerformance']
        print(f"ESG data for {ticker_symbol}:")
        print(f"  Total ESG Score: {esg_score.get('totalEsg', 'N/A')}")
        print(f"  Environment Score: {esg_score.get('environmentScore', 'N/A')}")
        print(f"  Social Score: {esg_score.get('socialScore', 'N/A')}")
        print(f"  Governance Score: {esg_score.get('governanceScore', 'N/A')}")
    else:
        print(f"ESG data not available for {ticker_symbol}")

    return esg_data

# Exemple d'utilisation
sah = get_esg_score("XYL")  # Remplace par le ticker que tu veux
```
